```python
import pandas as pd
import numpy as np
from pyorb import kepler
```

```python
df = pd.read_csv("sbdb_query_results.csv")
df[['i', 'om', 'w', 'ma']] = df[['i', 'om', 'w', 'ma']] * np.pi / 360
df['a'] = df['a'] * 1.495978707e8
```

```python
df['E'] = kepler.mean_to_eccentric(np.array(df['ma']), np.array(df['e']))
df['ta'] = 2 * np.arctan(np.sqrt((1 + df['e'])/(1 - df['e'])) * np.tan(df['E'] / 2))
```

```python
array = kepler.kep_to_cart(np.array([df['a'], df['e'], df['i'], df['w'], df['om'], df['ta']]), mu=9.90693e20, degrees=False)
df_cart = pd.DataFrame(array.T, columns=['x', 'y', 'z', 'vx', 'vy', 'vz'])

df_cart['spkid'] = df['spkid']
df_cart = df_cart[['spkid', 'x', 'y', 'z', 'vx', 'vy', 'vz']]
df_cart
```

```python
df_cart.to_csv("..\\Data\\centaurData.csv", index=False)
```
